# Step 1: Introduction

Wikipedia's definition of body fat percentage is the total mass of fat divided by total body mass, times 100. It is considered as an important measure of fitness level based on various theoretical approaches about the relationships between body fat percentage, health, athletic capacity, etc. There exist different measurement techniques, such as underwater weighing, whole-body air displacement plethysmography, body average density measurement and so on. However, none of them can achieve simplicity, robustness and precision simultaneously, which motivates our project on developing a new method to estimate percentage of body fat of men, that is easy-to-use, robust, scalable and accurate.

# Step 3: Statistical Modeling and Analysis

Since there is strong linear relationship between $Bodyfat$ and $\frac{1}{Density}$, we decide to first try to predict $Density$, and then find the corresponding $Bodyfat$.

Note that $\frac{1}{Density}=\frac{Volume}{Weight}$, where $Volume$ has closer relationship with our measurable quantitives such as $Weight$, $Hip$, $Wrist$, etc. Actually, we will show that the SLR has a very high $R^2$ which is above 0.997.

Hence, we gonna first fit a SLR for $Volume$ which can be calculated by $\frac{Weight}{Density}$. Then we get our $\frac{1}{Density}$ given by $\frac{Volume}{Weight}$. Finally, we get the predicted $Bodyfat$. The procedure is like the following workflow:

$$\text{Measurable variables}\Rightarrow\text{Volume}\Rightarrow\frac{1}{\text{Density}}\Rightarrow\text{Bodyfat}$$


The first step of this part is data manipulation. We think that the variables with unit $cm$ should take the third power so that their unit becomes $cm^3$. We believe that the power transformation would help to satisfy the linearity assumption.

```r
#### Data transformation####
bodyfat$NECK3 <- bodyfat$ADIPOSITY^3
bodyfat$CHEST3 <- bodyfat$CHEST^3
bodyfat$ABDOMEN3 <- bodyfat$ABDOMEN^3
bodyfat$HIP3 <- bodyfat$HIP^3
bodyfat$THIGH3 <- bodyfat$THIGH^3
bodyfat$KNEE3 <- bodyfat$KNEE^3
bodyfat$ANKLE3 <- bodyfat$ANKLE^3
bodyfat$BICEPS3 <- bodyfat$BICEPS^3
bodyfat$FOREARM3 <- bodyfat$FOREARM^3
bodyfat$WRIST3 <- bodyfat$WRIST^3
```

The second step is to do variable selection. To get less variables and make our rule-of-thumb as simple as possible, we choose BIC as our criterion in the stepwise variable selection procedure.

```r
#### Variable selection ####
bodyfat$volume <- bodyfat$WEIGHT/bodyfat$DENSITY
volume_data <- bodyfat
volume_data$IDNO<-NULL;volume_data$BODYFAT<-NULL;volume_data$DENSITY<-NULL
step_model <- step(lm(volume~.,data=volume_data),direction = "both",k=log(nrow(volume_data)),trace = 0)
summary(step_model)
```

Now we have the model of the first part, that is 

$$\text{Volume}_i=\beta_0+\beta_1\text{Weight}_i+\beta_2\text{Hip}_i+\beta_3\text{Wrist}_i+\beta_4\text{Abdomen}_i^3+\beta_5\text{Hip}_i^3+\varepsilon_i, \quad \varepsilon_i\sim N(0,\sigma^2)$$

And the fitted model is:

$$\text{Volume}=-1.206+0.920\text{Weight}+0.217\text{Hip}-0.250\text{Wrist}+6.105(\text{Abdomen}^3/10^6)-7.604(\text{Hip}^3/10^6)$$

For example, a man who is 70kg, hip circumference equals to 95cm, wrist circumference equals to 17cm, abdomen 2 circumference equals to 85cm would have volume around 65.93$dm^3$. There is a 95% probability that his volume is between 64.55$dm^3$ and 67.31$dm^3$.

```r
predict(step_model,newdata=data.frame(WEIGHT=70,HIP=95,WRIST=17,ABDOMEN3=85^3,HIP3=95^3),interval="predict")
```

**Observation 1**: Weight, Height, Hip circumference, Wrist circumference, abdomen 2 circumference explains about 99.7% of the variation in Volume.

**Observation 2**: Men gain 0.920$dm^3$ on average for every extra kilogram weight with other variables being the same.

**Observation 3**: Men gain 0.217$dm^3$ on average for every extra centimetre hip circumference with other variables being the same.

**Observation 4**: Men loss 0.250$dm^3$ on average for every extra centimetre wrist circumference with other variables being the same.

**Observation 5**: Men gain $6.105\times10^{-6}dm^3$ on average for every extra $cm^3$ abdomen 2 $\text{circumference}^3$ with other variables being the same.

**Observation 6**: Men loss $7.604\times10^{-6}dm^3$ on average for every extra $cm^3$ $\text{hip}^3$ with other variables being the same.

To test whether each variable has significant influence on volume, the null hypothesis for each $\beta_i$ should be $H_0:\beta_i=0$, and the alternative is $H_1:\beta_i\neq0$. The associated p-value for the 6 parameters are all less than 0.05. Hence, reject each $H_0$ with at least 95% confidence.
The 95% confidence interval for $\beta_i,i=1,2,...,5$ are, respectively, (0.895,0.944), (0.141,0.293), (-0.392,-0.108), (5.45$\times10^{-6}$,6.76$\times10^{-6}$), (-9.63$\times10^{-6}$,-5.58$\times10^{-6}$)

```r
confint(step_model)
```

Next, we need to fit the SLR between $Bodyfat$ and $\frac{1}{Density}$:

$$ \text{Bodyfat}_i = \beta_0 + \beta_1\frac{1}{\text{Density}}_{i} + \epsilon_i, \quad{} \epsilon_i \sim N(0,\sigma^2) $$

```r
bodyfat_density <- data.frame(BODYFAT=bodyfat$BODYFAT,DENSITY_INVERSE=1/bodyfat$DENSITY)
density_model <- lm(BODYFAT~DENSITY_INVERSE,bodyfat_density)
summary(density_model)
```

Now we have the model of the second part, that is

$$\text{Bodyfat}=-414+\frac{457}{\text{Density}}$$

For example, for a male with body density equals to 1.07$gm/cm^3$, his predicted bodyfat percentage would be around 12.90%. There is a 95% probability that his body fat is between 12.69% and 13.12%.

```r
predict(density_model,newdata=data.frame(DENSITY_INVERSE=1/1.07),interval="predict")
```

**Observation 1**: $\frac{1}{Density}$ explains about 99.98% of the variation in Volume (based on $R^2$).

**Observation 2**: Men gain 457% of bodyfat% on average for every extra unit $\frac{1}{Density}$.

We can almost surely declare that there is linear relationship between $\frac{1}{Density}$ and $Bodyfat$.

Hence, the final model would be:

$$\text{Bodyfat}=-414+(-5511+420\text{Weight}+99\text{Hip}-114\text{Wrist}+2790(\text{Abdomen}^3/10^6)-3475(\text{Hip}^3/10^6))/\text{Weight}$$

Back to the example in the SRL for volume. If we consider the relationship between $\frac{1}{Density}$ and $Bodyfat$ contains no variance. Then for a man who is 70kg, hip circumference equals to 95cm, wrist circumference equals to 17cm, abdomen 2 circumference equals to 85cm would have volume around 65.93$dm^3$, there is a 95% probability that his volume is between 64.55$dm^3$ and 67.31$dm^3$. His $\frac{1}{Density}=\frac{Volume}{Weight}$ should be around 0.942 and has 95% probability to fall in (0.922$cm^3/gm$,0.961$cm^3/gm$). The 95% confidence interval for his bodyfat% would be (7.48,25.35), around 16.41.


```r
predict(step_model,newdata=data.frame(WEIGHT=70,HIP=95,WRIST=17,ABDOMEN3=85^3,HIP3=95^3),interval="predict")/70*457-414
```

To check whether the model is doing the right thing with no over fitting, we decide to make K-fold cross validition.

Firstly, fit the model between bodyfat and density. Secondly, fit the model between volume and the measuable quantitives. In addition,predict the volume of testdata. At last, calculate predicted density values and then use them to predict bodyfat. 

```r
predict_bodyfat <- function(traindata,testdata){
  bodyfat_density_data_train <- data.frame(BODYFAT=traindata$BODYFAT,DENSITY_INVERSE=1/traindata$DENSITY)
  bodyfat_density_model <- lm(BODYFAT~DENSITY_INVERSE,bodyfat_density_data_train)
  volume_data_train <- traindata
  volume_data_train$volume <- volume_data_train$WEIGHT/volume_data_train$DENSITY
  volume_data_train$IDNO <- NULL;volume_data_train$BODYFAT <- NULL;volume_data_train$DENSITY <- NULL
  volume_model <- lm(volume~WEIGHT+HIP+WRIST+ABDOMEN3+HIP3,volume_data_train)
  volume_data_test <- testdata
  predicted_volume <- predict(volume_model,volume_data_test)
  bodyfat_density_data_test <- data.frame(BODYFAT=testdata$BODYFAT,DENSITY_INVERSE=predicted_volume/testdata$WEIGHT)
  predicted_bodyfat <- predict(bodyfat_density_model,bodyfat_density_data_test)
  return(predicted_bodyfat)
}
```

The K-fold cross validition gives MSE about 14.

```r
set.seed(2018)
MSE <- c()
for(i in 1:10){
  test_index <- sample(1:nrow(bodyfat),floor(nrow(bodyfat)/10))
  testdata <- bodyfat[test_index,];traindata <- bodyfat[-test_index,]
  predicted.values <- predict_bodyfat(traindata,testdata)
  MSE <- c(MSE,sum((predicted.values-testdata$BODYFAT)^2)/floor(nrow(bodyfat)/10))
}
mean(MSE)
```

# Step 4: Diagnostics

This part mainly check whether the requirements for SLR of the volume model are satisfied. There are four assumptions to check:

**Linearity**: Bodyfat % has linear relationship with those explanatory variables.

**Homoskedastic**: The variance on the error terms is assumed to be constant.

**Normality**: The error terms are indipendently normal distributed with mean equals to 0.

**Additive**: The explanatory variables are additive, meaning that they are independent with each other and do not have interaction effect.

Meanwhile, this part would detect outliers and influencial points.

### First 
To check **Linearity** and **Homoskedastic**, we gonna look at the residual plot:

```r
plot(predict(step_model),rstandard(step_model),pch=23,bg="red",cex=1.2,
    xlab="Predicted Volume(dm^3)", ylab="Standardized Residuals",main="Standardized Residual Plot")
abline(a=0,b=0,col="black",lwd=3)
```

The standard errors are symmetrically distributed around 0. There is no obvious heteroscedasticity pattern. There is an outlier, observation No.39, with predicted volume over 160. This man has body volume 161$dm^3$. This obsercation has a extremely high Pii, which means its X data is far away from others. However, the observation does not has a high cook's distance. Hence, the observation will not be excluded.

There is also a data point which has a very high cook's distance. The observation number is 250. This men looks normal and the Pii is not high. Hence, the observation will not be excluded.

Overall, the **Linearity** and **Homoskedastic** assumptions hold.

```r
bodyfat[predict(step_model)>160,c(1,ncol(bodyfat),2:17)]
```

```r
pii = hatvalues(step_model)
cooki = cooks.distance(step_model)
par(mfrow = c(2,1))
n = dim(bodyfat)[1]
plot(1:n,pii,type="p",pch=23,bg="red",cex=1.2,
     xlab="Index (Each Observation)",ylab="Pii",main="Influence Values (Pii)")
plot(1:n,cooki,type="p",pch=23,bg="red",cex=1.2,
     xlab="Index (Each Observation)",ylab="Cook's Distance",main="Influence Values (Cook's Distance)")
```

```r
bodyfat[cooki>0.08,c(1,ncol(bodyfat),2:17)]
```

### Third

To further check **additive** and **linearity** assumption, we need to look at the Scatterplot of each variable verses the standard residual and volume.

# Step 5: Conclusion

Our Proposed Model to predict bodyfat %:

$$\text{Bodyfat}=(-5500+6\text{Weight}+1000\text{Hip}-1000\text{Wrist}+3\text{Abdomen}^3-3.5\text{Hip}^3)/\text{Weight}$$

**Rule-of-thumb**: 6 times weight(kg) plus 1000 times hip circumference(dm) minus 1000 times wrist(dm) plus 3 times abdomen 2 circumference(dm) cube minus 3.5 times hip circumference(dm) cube minus 5500 over weight(kg).

**Example Usage**: For example, for a male of weight 70 kg, hip circumference 10 dm, abdomen 9 dm, wrist 1.7 dm, with the proposed rule-of-thumb, the estimated body fat is about 27%.

**Inference about Relationship**: There is strong linear relationship between bodyfat % and the explainary variables . This relationship is significant at the $\alpha = 0.05$ level for all the parameters. The explainary variables explains about 99.7% of all the variation in bodyfat %.

**Strengths and Weakness of Model**:

  1. Linearity: All the exlainary varaibles have strong linear relationship with the response varaibles in both two models.
  2. Additivity: In the volume model, most variables can be added to each other.
  3. Constant effects: since we don't split the whole dataset into different groups, our model may not be very precise for specific groups.
  4. Fixed X: This seems reasonable for this data.
  5. Normally distributed errors: after our data cleaning procedure, the QQ plot of the model looks perfect.
  6. Constant variance: satisfied by looking at the residual plot.
  7. The prediction of each model of the two steps are both very precise.
  8. The final model is complicated in calculation.

Overall, our model provides precise prediction of bodyfat % based on weight, hip circumference, wrist circumference and abdomen circumference although it is a little complicated.
